```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table

# Datos de conexión
username = 'edgar'
password = 'mucikaa12'
server = 'localhost'
database = 'accidentes'

# Crear motor de SQLAlchemy
engine = create_engine(f'mysql+pymysql://{username}:{password}@{server}/{database}')

# Leer el archivo Excel
df = pd.read_excel('OaxacaDatosDias.xlsx')

# Crear un objeto MetaData
metadata = MetaData()

# Reflejar las tablas existentes en el objeto MetaData
metadata.reflect(engine)

# Obtener referencias a las tablas
tiempo = metadata.tables['tiempo']
ubicacion = metadata.tables['ubicacion']
vehiculo = metadata.tables['vehiculo']
causa_accidente = metadata.tables['causa_accidente']
victima = metadata.tables['victima']
accidente = metadata.tables['accidente']

# Procesar y registrar los datos en las tablas correspondientes
with engine.begin() as connection:  # Usar el contexto de conexión
    for _, row in df.iterrows():
        # Insertar en la tabla tiempo
        tiempo_insert = {
            'anio': row['ANIO'],
            'mes': row['MES'],
            'id_hora': row['ID_HORA'],
            'id_minuto': row['ID_MINUTO'],
            'id_dia': row['ID_DIA'],
            'diasemana': row['DIASEMANA'],
            'dia_fes': row['DIA_FES'],
            'fin_semana_largo': row['Fin_Semana_Largo']
        }
        result = connection.execute(tiempo.insert().values(tiempo_insert))
        tiempo_id = result.inserted_primary_key[0]

        # Insertar en la tabla ubicacion
        ubicacion_insert = {
            'urbana': row['URBANA'],
            'suburbana': row['SUBURBANA'],
            'nom_municipio': row['NOM_MUNICIPIO']
        }
        result = connection.execute(ubicacion.insert().values(ubicacion_insert))
        ubicacion_id = result.inserted_primary_key[0]

        # Insertar en la tabla vehiculo
        vehiculo_insert = {
            'tipaccid': row['TIPACCID'],
            'automovil': row['AUTOMOVIL'],
            'campasaj': row['CAMPASAJ'],
            'microbus': row['MICROBUS'],
            'pascamion': row['PASCAMION'],
            'omnibus': row['OMNIBUS'],
            'tranvia': row['TRANVIA'],
            'camioneta': row['CAMIONETA'],
            'camion': row['CAMION'],
            'tractor': row['TRACTOR'],
            'ferrocarril': row['FERROCARRI'],
            'motociclet': row['MOTOCICLET'],
            'bicicleta': row['BICICLETA'],
            'otrovehic': row['OTROVEHIC']
        }
        result = connection.execute(vehiculo.insert().values(vehiculo_insert))
        vehiculo_id = result.inserted_primary_key[0]

        # Insertar en la tabla causa_accidente
        causa_accidente_insert = {
            'causaacci': row['CAUSAACCI'],
            'caparod': row['CAPAROD']
        }
        result = connection.execute(causa_accidente.insert().values(causa_accidente_insert))
        causa_accidente_id = result.inserted_primary_key[0]

        # Insertar en la tabla victima
        victima_insert = {
            'id_edad': row['ID_EDAD'],
            'condmuerto': row['CONDMUERTO'],
            'condherido': row['CONDHERIDO'],
            'pasamuerto': row['PASAMUERTO'],
            'pasaherido': row['PASAHERIDO'],
            'peatmuerto': row['PEATMUERTO'],
            'peatherido': row['PEATHERIDO'],
            'ciclmuerto': row['CICLMUERTO'],
            'ciclherido': row['CICLHERIDO'],
            'otromuerto': row['OTROMUERTO'],
            'otroherido': row['OTROHERIDO'],
            'nemuerto': row['NEMUERTO'],
            'neherido': row['NEHERIDO']
        }
        result = connection.execute(victima.insert().values(victima_insert))
        victima_id = result.inserted_primary_key[0]

        # Insertar en la tabla accidente
        accidente_insert = {
            'id_tiempo': tiempo_id,
            'id_ubicacion': ubicacion_id,
            'id_vehiculo': vehiculo_id,
            'id_causa': causa_accidente_id,
            'id_victima': victima_id,
            'clasacc': row['CLASACC']
        }
        connection.execute(accidente.insert().values(accidente_insert))
```
